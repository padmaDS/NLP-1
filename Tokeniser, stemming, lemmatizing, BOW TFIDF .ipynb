```python
import nltk
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
#tokenisation converts paragraph in to sentenses and words
nltk.download('punkt')
sentenses = nltk.sent_tokenize(paragraph)
```

```python
sentenses
```

```python
#stemmer gives the base root word

stemmer = PorterStemmer()
```

```python
stemmer.stem('flying')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
# lemmatizer gives full meaning of the word

lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('sleep')
```

```python
len(sentenses)
```

```python
import re
corpus=[]
for i in range(len(sentenses)):
    review = re.sub('[^a-zA-Z]',' ', sentenses[i])
    review = review.lower()
    corpus.append(review)
    
```

```python
corpus
```

```python
stopwords.words('english')
```

```python
#stemming
for i in corpus:
    words=nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word))
```

```python
#lemmatizing
for i in corpus:
    words=nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemmatizer.lemmatize(word))
```
